# Police Crime Data & Population

### Combining crime data with population to get the crime rate

```python
# setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
```

## Setting up DataFrames

```python
#grab csv files for population and crime info
population_df = pd.read_csv("census_data.csv") 
police_df = pd.read_csv("CSV_FILES/PD_Crime_Data/ALL_CRIME.csv") 
```

```python
#set up population for merging
population_df = population_df.drop(["State Number"],axis=1)
population_df = population_df.drop(["place"],axis=1)
population_df["State"] = ["LA","MO", "OR","MA","TN","VA","CA","CO","SD","NV","IL","TX"]
population_df
```

```python
#check format before merging
police_df
```

```python
#merge on common column of state
pop_police_df = pd.merge(police_df, population_df, on="State", how="outer")
pop_police_df
```

```python
#reset index to the police department name
pop_police_df = pop_police_df.set_index("Unnamed: 0")
pop_police_df.index.name = None
pop_police_df
```

```python
#sort columns so the city,state & pop are on the left
pop_police_df = pop_police_df[['State', 'ORI', 'City, State', 'Population', 'Agg Assault', 'Arson', 'Auto Theft',
                               'Burglary', 'Homicide', 'Larceny','Property Crime', 'Robbery', 'Sex Crimes',
                               'Violent Crime', 'Total Crime', 'Total Person Crime', 'Total Property Crime']]

#make csv copy
pop_police_df.to_csv("CSV_Files/Crime_Rates/pd_and_pop.csv", encoding="utf-8")
pop_police_df
```

# Calculating Crime Rate

Crime rate calculation
source: https://oag.ca.gov/sites/all/files/agweb/pdfs/cjsc/prof10/formulas.pdf
"CRIME RATE – A crime rate describes the number of crimes reported to law enforcement agencies per 100,000 total population.  A crime rate is calculated by dividing the number of reported crimes by the total population; the result is multiplied by 100,000.  For example, in 2010 there were 58,100 robberies in California and the population was 38,826,898.  This equals a robbery crime rate of 149.6 per 100,000 general population."

## All Crime Rates 
### All crime columns have crime rate calculated and added as sepate columns

```python
#making crime rate for all categories
#very large does not show all columns accross
all_crime_rates = pd.DataFrame(pop_police_df[['State', 'ORI', 'City, State', 'Population', 'Agg Assault', 'Arson', 'Auto Theft',
                               'Burglary', 'Homicide', 'Larceny','Property Crime', 'Robbery', 'Sex Crimes',
                               'Violent Crime', 'Total Crime', 'Total Person Crime', 'Total Property Crime']])

all_crime_rates["Agg Assault Crime Rate"] = (all_crime_rates["Agg Assault"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Arson Crime Rate"] = (all_crime_rates["Arson"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Auto Theft Crime Rate"] = (all_crime_rates["Auto Theft"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Burglary Crime Rate"] = (all_crime_rates["Burglary"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Homicide Crime Rate"] = (all_crime_rates["Homicide"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Larceny Crime Rate"] = (all_crime_rates["Larceny"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Property Crime Rate"] = (all_crime_rates["Property Crime"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Robbery Crime Rate"] = (all_crime_rates["Robbery"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Sex Crimes Rate"] = (all_crime_rates["Sex Crimes"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Violent Crime Rate"] = (all_crime_rates["Violent Crime"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Total Crime Rate"] = (all_crime_rates["Total Crime"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Total Person Crime Rate"] = (all_crime_rates["Total Person Crime"] / all_crime_rates["Population"]) * 100000
all_crime_rates["Total Property Crime Rate"] = (all_crime_rates["Total Property Crime"] / all_crime_rates["Population"]) * 100000

#to csv file
all_crime_rates.to_csv("CSV_Files/Crime_Rates/all_crime_rates.csv", encoding="utf-8")
all_crime_rates
```
