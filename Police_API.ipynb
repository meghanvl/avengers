```python

#google coord
#city_list = ["Kansas City, MO
#    "New Orleans, LA 
#    "Dallas, TX 
#    "Portland, OR 
#    "Charlottesville, VA  - https://www.lat-long.com/Latitude-Longitude-1498463-Virginia-Charlottesville.html
#    "Chicago, IL
#    "Denver, CO 
#    "Sioux Falls, SD
#    "Las Vegas, NV 
#    "San Francisco, CA
#    "Chattanouga, TN
#    "Boston, MA ]

#testcoord_list = ["39.952583,-75.165228", "40.712772, -74.006058"]
#lat and long source: https://latitudelongitude.org/us/
coord_list = ["39.09973, -94.57857",
              "29.95465, -90.07507",
              "32.78306, -96.80667",
              "45.52345, -122.67621",
              "38.0293059, -78.4766781",
              "41.85003, -87.65005",
              "39.73915, -104.9847",
              "43.54997, -96.70033",
              "36.17497, -115.13722",
              "37.77493, -122.41942",
              "35.04563, -85.30968",
              "42.35843, -71.05977"  ]
name = []
lat = []
lng = []

for coords in coord_list:
    target_coordinates = coords
    target_search = "Police"
    target_radius = 8000
    target_type = "Police"

# set up a parameters dictionary
    params = {"location": target_coordinates,
              "keyword": target_search,
              "radius": target_radius,
              "type": target_type,
              "key": gkey}

    # base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # run a request using our params dictionary
    response = requests.get(base_url, params=params)
     # convert response to json
    places_data = response.json()
    # Print the name and address of the first police that appears
    #print(places_data)
    name.append(places_data["results"][0]["name"])
    #print(places_data["results"][0]["vicinity"])
    lat.append(places_data["results"][0]["geometry"]["location"]["lat"])
    lng.append(places_data["results"][0]["geometry"]["location"]["lng"])
    #create dataframe
    
```

```python
# print the response url, avoid doing for public github repos in order to avoid exposing key
print(response.url)
coord_df = pd.DataFrame({"name" : name,
                       "lat" : lat,
                       "lng" : lng})
coord_df
```
